```python

# color,marker,linewidth,axis,datasets,and the title, axis, grid


import numpy as np

# x = np.array([1,2,3,4])

x = np.arange(0,5,0.1)
y = x**3

y2 = x**2


# plt.plot(x,y,color='blue',marker='+')

plt.plot(x,y,color='black',linewidth=4,label='x^3')

plt.plot(x,y2,color='red',linewidth=4,label='x^2')

# plt.axis([0,10,0,200])

plt.ylabel('Y')
plt.xlabel('X')

plt.grid()
plt.legend()

plt.text(2,80,'Text',fontsize = 15)

plt.title('Matplotlib-demo')

plt.show()
```

```python

# Bubble chart

year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.scatter(year,salary,s=population,c='r',alpha=0.5) # s is size which is equal to an array meaning diff values for each term

plt.grid()
plt.title('bubble chart')
plt.xlabel('year')
plt.ylabel('population')

plt.show()

```

```python

import matplotlib.pyplot as plt
import numpy as np
employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]

employees = np.array(employees)
year = np.array(year)
revenue = np.array(revenue)

count = np.arange(len(employees))

plt.scatter(year,revenue,s=employees,c=count,alpha=0.9,edgecolor='black',marker='^')

for i in range(len(employees)):
    
    plt.text(year[i]+0.1,revenue[i]+0.1,employees[i])

plt.grid()
plt.show()

index = np.where(year==2008)
# print(index)

print(year[index[0][0]],revenue[index[0][0]],employees[index[0][0]])

index2 = np.where(year==2015)

print(year[index2[0][0]],revenue[index2[0][0]],employees[index2[0][0]])
```

```python
# pie graph

labels = ['A','B','C','D']
sizes = [3,4,6,2]

color = ['blue','purple','red','pink']

explode = [0.1,0,0,0]

plt.pie(sizes,colors=color,explode=explode,labels=labels,autopct='%.1f%%',counterclock = False,startangle=250)
plt.title('split among classes')
plt.axis('equal')
plt.show()
```
